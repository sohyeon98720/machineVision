## 1) Check contrast

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# 이미지 파일 읽어오기
img = cv2.imread('landscape.jpg') # cv2.IMREAD_COLOR (default) / cv2.IMREAD_GRAYSCALE / cv2.IMREAD_UNCHANGED
# img = cv2.imread('lung.jpg')

# 히스토그램 생성하기
hist, bins = np.histogram(img, 256, [0,256])

# 정규화된 누적 히스토그램 생성하기
cdf = hist.cumsum()
cdf_normalized = cdf * hist.max() / cdf.max() # 정규화 해주는 이유 => 히스토그램과 같은 스케일로 표현하기 위함

# 히스토그램 시각화하기
plt.plot(cdf_normalized, color='b')
plt.hist(img.flatten(), 256, [0,256], color='r')
plt.xlim([0,256])
plt.legend(['cdf', 'histogram'], loc='upper left')
plt.show()

# 이미지 창 띄우기
cv2.imshow('image', img)
cv2.waitKey()

cv2.destroyAllWindows()
```

## 2) Min-max stretching

- 가장 작은 값을 0, 가장 큰 값이 255가 되도록 매핑시킴

\begin{align}
X_{new} = \frac{X_{input} - X_{min}}{X_{max} - X_{min}} \times 255
\end{align}

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# 이미지 파일 읽어오기
img = cv2.imread('landscape.jpg')

# min-max stretching 수행하기
img_minmax = (img - img.min()) / (img.max() - img.min()) * 255.0
img_minmax = img_minmax.astype(np.uint8)

# 히스토그램 생성하기
hist, bins = np.histogram(img_minmax, 256, [0,256])

# 정규화된 누적 히스토그램 생성하기
cdf = hist.cumsum()
cdf_normalized = cdf * hist.max() / cdf.max()

# 히스토그램 시각화하기
plt.plot(cdf_normalized, color='b')
plt.hist(img_minmax.flatten(), 256, [0,256], color='r')
plt.xlim([0,256])
plt.legend(['cdf', 'histogram'], loc='upper left')
plt.show()

# 이미지 창 띄우기
cv2.imshow('image', img_minmax)
cv2.waitKey()

cv2.destroyAllWindows()
```

## 3) Percentile stretching

```python
# 과제부분

import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def percentile_stretching(img, p_low, p_high):
    img_ptile=img;
    #print(img.dtype)
    if(p_low<=p_high):
        low=0;high=0;
        n_low=p_low * img.size /100;
        n_high=(100-p_high) * img.size /100;###퍼센트에 해당하는 픽셀의 갯수
        #print(n_low,n_high);
        hist, bins = np.histogram(img, 256, [0,256])###히스토생성
        
        cnt=0;
        for i in range(0,256):
            cnt+=hist[i];
            if(cnt>n_low):
                low=i;
                print("low: ",low);
                break;
            
        ###low값 찾기
        
        cnt=0;
        for i in range(255,-1,-1):
            cnt+=hist[i];
            if(cnt>n_high):
                high=i;
                print("high: ",high);
                break;
            
        ##high값 찾기
        
        for i in range(0,img.shape[0]):
            for j in range(0,img.shape[1]):
                #print(img[i,j]);
                if(img[i,j]<low):
                    img_ptile[i,j]=0;
                elif(img[i,j]>high):
                    img_ptile[i,j]=255;
                else: img_ptile[i,j]=(img[i,j]-low)/(high-low) * 255.0;
                
        ###low보다 왼쪽에 위치하면 0, high보다 오른쪽에 위치하면 255, 나머지 스트레칭
        print(img_ptile.dtype)
        ###이미 데이터타입이 uint8이라서 밑의 줄 수행안함
        #img_ptile = img_ptile.astype(np.uint8) ###unsigned int 8비트 짜리로 형변환
        
        return img_ptile

    else :###인자를 잘못 입력한 경우(p_low>p_high) 두 인자를 바꿔 다시 수행
        percentile_stretching(img,p_high,p_low);


# 이미지 파일 읽어오기
img = cv2.imread('landscape.jpg',cv2.IMREAD_GRAYSCALE)

# percentile stretching 수행하기
img_percentile = percentile_stretching(img, 3, 97)

# # # 히스토그램 생성하기
hist, bins = np.histogram(img_percentile, 256, [0,256])

# # # 정규화된 누적 히스토그램 생성하기
cdf = hist.cumsum()
cdf_normalized = cdf * hist.max() / cdf.max()

# # # 히스토그램 시각화하기
plt.plot(cdf_normalized, color='b')
plt.hist(img_percentile.flatten(), 256, [0,256], color='r')
plt.xlim([0,256])
plt.legend(['cdf', 'histogram'], loc='upper left')
plt.show()

# # # 이미지 창 띄우기
cv2.imshow('image', img_percentile)
cv2.waitKey()

cv2.destroyAllWindows()
```

## 4) Histogram equalization

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# 이미지 파일 읽어오기
img = cv2.imread('landscape.jpg')

# 히스토그램 평활화 수행하기
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
equalized = cv2.equalizeHist(gray)
img_equalized = cv2.merge([equalized, equalized, equalized])

# 히스토그램 생성하기
hist, bins = np.histogram(img_equalized, 256, [0,256])

# 정규화된 누적 히스토그램 생성하기
cdf = hist.cumsum()
cdf_normalized = cdf * hist.max() / cdf.max()

# 히스토그램 시각화하기
plt.plot(cdf_normalized, color='b')
plt.hist(img_equalized.flatten(), 256, [0,256], color='r')
plt.xlim([0,256])
plt.legend(['cdf', 'histogram'], loc='upper left')
plt.show()

# 이미지 창 띄우기
cv2.imshow('image', img_equalized)
cv2.waitKey()

cv2.destroyAllWindows()
```
